```python
# Lendo os dados de contribuição ao PIB por setor da economia
#df = pd.read_csv('drive/MyDrive/ASA 2023/Colab Notebooks/data/pib2.csv')4
df = pd.read_csv('data/pib2.csv')
df
```

## Linha

```python
df.columns
```

```python
filtered_df = df[df["Ano"] >= 2015]
filtered_df
```

```python
filtered_2019 = df[df["Ano"] == 2019]
filtered_2019
```

```python
# Como fazer um gráfico de linha simples usando plotly express
px.line(data_frame=filtered_2019, x='Trimestre', y='Indústrias de transformação')
```

```python
# Adicionando título e pontos
px.line(data_frame=filtered_2019, x='Trimestre', y='Indústrias de transformação', title='Valor de contribuição para o PIB por trimestre', markers=True)
```

```python
# Exemplo de plot com multiplas linhas
# fig.update_layout() é usado para modificar todas as características do gráfico
fig = px.line(data_frame=filtered_df, x='Trimestre', y='Indústrias de transformação', title='Valor de contribuição para o PIB por trimestre', markers=True, color='Ano')
fig.update_xaxes(type='category')
fig.show()
```

## Dispersão

```python
#Relembrando como é o df
df
```

```python
  #Scatter plot pode nos ajudar a entender correlações
scatter_fig = px.scatter(data_frame=df, x='Indústrias extrativas', y='Indústrias de transformação', height=500, width=700, trendline="ols")
scatter_fig.show()
```

## Barra

```python
#Relembrando como é o df
df
```

```python
#Agrupando os valores pelo ano
df_agrupado = df.groupby('Ano').sum().drop(columns='Trimestre')
df_agrupado
```

```python
px.bar(data_frame=df_agrupado.loc[2010])
```

## Exercício 2

Exiba em um gráfico de barras para os setores de 'Comércio' e 'Construção' no ano de 2015:

```python
#Seu código aqui
px.bar(data_frame=df_agrupado.loc[[2015, 2016], ['Comércio', 'Construção']], barmode='group')
```

```python
#Gráfico de barras horizontal
px.bar(data_frame=df_agrupado.loc[2020], orientation='h')
```

```python
#Exemplo de stacked bar
px.bar(data_frame=df_agrupado)
```

## Histograma

```python
hist_df = df[['Indústrias extrativas', 'Indústrias de transformação']]
px.histogram(data_frame=hist_df, nbins=40)
```

## Formatação de gráfico

```python
#Relembrando o gráfico de dispersão
scatter_fig
```

```python
#Adicionando algumas opções de formatação
scatter_fig.update_layout(
    title='Brincando com a formatação',
)
scatter_fig.update_traces(
    marker={'size': 12, 'color': 'violet'}
)
scatter_fig.update_xaxes(
    type="log",
    range=[3,5] # log range: 10^3=1000, 10^5=100000,
)
```
