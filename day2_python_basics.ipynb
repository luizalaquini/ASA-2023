```python
# função com argumento opcional

def media(x, y, tipo="aritmetica"):
    """
    Calcula a média de um par de números.

    Parameters
    ----------
    x : int or float
        Primeiro número.
    y : int or float
        Segundo número.
    tipo : str, default 'aritmetica'
        Tipo de média a ser calculado, que pode ser 'aritmetica' ou 'geometrica'.

    Returns
    -------
    float
        A média entre ``x`` e ``y``.
    """
    # acima, vemos o exemplo de uma doc-string, que documenta a função

    if tipo == "aritmetica":
        return (x + y) / 2
    elif tipo == "geometrica":
        return (x * y) ** 0.5
```

```python
# por padrão, o tipo considerado é "aritmetica"
media(1, 2)
```

```python
# isso pode ser passado explicitamente, e o resultado é o mesmo
media(1, 2, tipo="aritmetica")
```

```python
# apenas se o tipo for alterado para "geometrica" o resultado muda
media(1, 2, tipo="geometrica")
```
